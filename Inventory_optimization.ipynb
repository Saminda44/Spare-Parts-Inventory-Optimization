```python
import pandas as pd

# Load the Excel file
file_path = "Jan 1st week Spare Parts PO.xlsx"
df = pd.read_excel(file_path)

# Convert date columns to datetime (errors='coerce' will turn invalid dates into NaT)
df['Created On'] = pd.to_datetime(df['Created On'], errors='coerce')
df['Goods Issue Date'] = pd.to_datetime(df['Goods Issue Date'], errors='coerce')

# Calculate Lead Time (in days)
df['Lead Time'] = (df['Goods Issue Date'] - df['Created On']).dt.days

# Ensure quantity columns are numeric
df['Order Quantity (Item)'] = pd.to_numeric(df['Order Quantity (Item)'], errors='coerce')
df['Confirmed Quantity (Item)'] = pd.to_numeric(df['Confirmed Quantity (Item)'], errors='coerce')

# Calculate Lost Qty
df['Lost Qty'] = df['Order Quantity (Item)'] - df['Confirmed Quantity (Item)']

df.head(10)
```
